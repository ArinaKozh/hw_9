```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('/content/Mall_Customers.csv')
```

```python
df
```

```python
print(df['Age'].mean())
```

```python
print(df['Spending Score (1-100)'].mean())
```

```python
sns.scatterplot(data = df, x='Spending Score (1-100)', y='Annual Income (k$)',hue='Age')
```

Наблюдается, что уровень трат не зависит от годового заработка, а у большинства людей уровень трат находится в диапазоне 40-60

```python
sns.relplot(x='Annual Income (k$)', y="Spending Score (1-100)", kind="line", data=df)
```

Самый большой скачок потребления приходится на население с заработком 70к$ в год

```python
sns.histplot(data = df, x = 'Age') 
```

Основная масса клиентов находится в диапазоне 30-35 лет
